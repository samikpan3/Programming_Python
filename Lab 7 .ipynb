Q2: 
For each combination of age and education, find how many people earn more than  50k 


```python
#Created a new df that only includes dataset where label is >50k. Then grouped this dataframe 
adult_50k = adult.loc[adult["label"] == ">50K"]
adult_50k.groupby(["age", "education"])["label"].count()
#The following table counts based on age and education, how many people earn more than 50k0 
```

Q3. 
Pivot the education and for each occupation find people earning more than 50k for each gender 

```python
#Pivot dataset:
grouped_50k = adult[adult["label"] == ">50K"]
grouped_50k.pivot_table("label", index = "occupation", columns = "sex", aggfunc = "size")
```

```python
user = pd.read_csv("u.user.txt", sep = "|")  #Loading the dataset
user.head()
```

q1


```python
mean_age = user["age"].groupby(user["occupation"]).mean()
print(mean_age)
```

q2. 
FInd female ratio per occupations and sort it from the most to least 

```python
female_count = user.groupby("occupation")["gender"].value_counts()  #Counting number of females and males grouped by occupation.
# print(female_count)
total_gender = user.groupby("occupation")["gender"].size() #Counting total number of people in each occupation
percent = (female_count/total_gender)*100  #Calculating the ratio
#percent.head()
occupationbygender_ratio = percent.to_frame().unstack() #Need to convert the percent which is in series form to dataframe for further calculations
#unstack the percent ratio dataframe so that male and female now in different columns


occupationbygender_ratio = occupationbygender_ratio.droplevel(0, axis=1) #drop the index "gender" in position 1 through column to extract female occupation_gender ratio
occupation_female_ratio =  occupationbygender_ratio.drop("M", axis = 1 )
#occupation_female_ratio.head()

occupation_female_ratio = occupation_female_ratio.sort_values('F', ascending=False)
print(occupation_female_ratio)

#occupation_female_ratio.info()
```

q3: For each occupation, calculate the min and max age

```python
user.groupby("occupation")["age"].agg(["min", "max"])  #Groupby occupation and use the column age. Use function min and max
```

q4: for combination of occupation and gender, calculate mean age

```python
user.groupby(["occupation", "gender"])["age"].mean().unstack() #Groupby two categories, pass them as list and use column age to calculate the mean
```
