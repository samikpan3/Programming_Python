How much was the revenue for the period in the dataset?

```python
#Revunue = qunatity * item_price 
data[["item_price", "quantity"]].head(5) #All items in the item price has a $ so we cnanot do calculation yet.  
#Removing the $ sign
```

```python
data["item_price"] = data["item_price"].str.replace('$' , '')  #Removing the $ for calculation
#data.head() 
data["item_price"] = data["item_price"].astype(str).astype(float)  #convert the string to float
data["quantity"] = data["quantity"].astype(float) #convery quantity to float to make calculation easier


```

```python
data.info() #checking if data type is similar
```

```python
data['Revenue'] = data["quantity"] * data["item_price"]  #Revenue = quantitiy * item price
data['Revenue'].sum()


```

we can see that the total revenue collected is 39237.02

Find the orders that cost more than $100

```python
data["Total_cost"] = data["item_price"] * data["quantity"] 
data.info()
```

```python
total = data[["Total_cost", "order_id"]].groupby(["order_id"]).sum()  #groupby.orderid groupby order id did the sum
total_100 = total[total["Total_cost"] > 100] 
print(total_100)   #printing the order_id whose total cose was greater than 100. 
```

## Task 2 

```python
data_student = pd.read_csv("student-mat.txt" , sep = ',') #Loading the dataset.
data_student.head()
```

Slicing the dataframe from ‘school’ until ‘guardian’ and save it into new dataframe

```python
df_student2 = data_student.loc[:,'school':'guardian'] #Creating a new dataframe
df_student2
```

```python
f = lambda x: x.str.title()  #Creating a function that converts the first letter into a capital
df_student2[["Mjob", "Fjob"]] = df_student2[["Mjob", "Fjob"]].apply(f)
df_student2.head(10)
```

Add a new column "legal_drinker" which has a boolean value. If age is below 21, the value is 
false; otherwise, the value is true. 

```python
df_student2["legal_drinker"] = np.where(df_student2["age"] < 21, False, True)  #pass true for age greater or bigger than 21
df_student2.head(10)
```
